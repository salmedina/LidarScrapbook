```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
import numpy as np

%matplotlib notebook
color_palette = sns.color_palette("Set2", n_colors=4)
```

Data acquisition

```python
data = np.array([(-1.6862784794398717, -1.408254895891462, -0.057691804000309536),
(-1.3169442749023437, -2.451765441894531, -0.042917299270629886),
(-1.506814684186663, -1.5591510591052828, -0.21929232279459634),
(-0.4909800479286595, -2.9155048571134867, -0.29087892331575094),
(-1.5634245872497559, -1.7947829961776733, -0.05750056356191635),
(-0.6907578150431315, -2.7593699137369794, -0.23650180498758952),
(0.22266177336374918, -2.7776126861572266, 0.0887807806332906),
(-0.4501505851745605, -3.008653259277344, 0.1329094409942627),
(0.7466692924499512, -2.489933490753174, -0.454537570476532),
(0.18010187149047852, -2.682621955871582, -0.3347104787826538),
(-0.7525075912475586, -2.8667579650878907, 0.15569522380828857),
(-1.7614749908447265, -1.6392866134643556, 0.1264027118682861)])

xs = data[:, 0]
ys = data[:, 1]
zs = data[:, 2]
```

### Clustering

```python
clustering = DBSCAN(metric='l2', min_samples=3, eps=0.7).fit(data)
clusters = clustering.labels_ + 1
print(clusters)
```

### Plotting

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

colors = [color_palette[i] for i in clusters]
ax.scatter(xs, ys, zs, c=colors, marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# rotate the axes and update
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
    plt.show()
```
